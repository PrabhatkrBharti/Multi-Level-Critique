```python
import json
import numpy as np 
import pandas as pd 
import csv
import pickle
import matplotlib.pyplot as plt
import tensorflow as tf
import pickle
```

```python
## PATH TO OUR DATASET ##
PATH = 'dataset.csv' ## ADD PATH HERE
df = pd.read_csv(PATH)
df['count'] = np.ones(df.shape[0])
df = df[['Tone', 'N/C', 'count']]
df
```

```python
# GROUP BY
df.groupby(['N/C', 'Tone']).count()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
df.groupby(['N/C', 'Tone']).count().unstack().plot(kind = 'bar', ax=ax)
plt.legend(['Tone Level 1', 'Tone Level 2', 'Tone Level 3'], fontsize=16)
plt.xlabel("Levels of Constructiveness with Politeness Tone sub-levels" ,fontsize=16)
plt.ylabel("Number of Reviews", fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14, rotation=0)
plt.show()
```
